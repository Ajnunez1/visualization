####Lecture 3: Reproducible Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import PIL
import requests
```

```python
#setting the seed for reproducibility
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
### Plotting two lines on the same graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
#Adding a legend to differentiate the two lines
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

```python
# Customizing the legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right'
,
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns
shadow = True )
fig.show()
```

```python
#moving the legend outside of the plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
          bbox_to_anchor =(1, 1))
fig.show()
```

```python
#Let’s start by putting our data into a scatter plot this time
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95,"This value is important!")
fig.show()
```

```python
#modifying the text anotation
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95,"This value is important!",
        ha='center', #horizontal alignment
        color='red', #text color
        size=20) #text size
fig.show()
```

```python
#####Positioning text annotations

fig, ax = plt.subplots(figsize=(5, 3))
ax.axis([0, 10, 0, 10])


ax.text(1, 5, "Data:(1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, "Axes:(0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, "Figure:(0.2, 0.2)", transform=fig.transFigure)

plt.show()
```

```python
#arrow annotation
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!'
, xy=(10, 95), xytext=(20, 94),
arrowprops=dict(facecolor='black'))
fig.show()
```

```python
#arrow annotation
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', 
            xy=(10, 95), 
            xytext=(20, 94),
            arrowprops=dict(arrowstyle="wedge", color="hotpink"))
fig.show()
```

```python
#Removing tick marks and labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator()) #Use locator to remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter())#Use formatter to remove tick labels
```

####Lecture 4: Subplots and Combining
Visualizations

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
#Introducing subplots

fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))
ax1.scatter(x1, y1) #Scatter plot of the first dataset
ax2.bar(x2, y2) #Bar plot of the second dataset
fig.show()
```

```python
#Customizing subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))
ax1.scatter(x1, y1, c='indigo', marker='*', s=100) #Scatter plot of the first dataset
ax2.bar(x2, y2, color='hotpink') #Bar plot of the second dataset
ax1.set_xlabel('Axis 1', fontdict=font1)
ax2.set_xlabel('Axis 2', fontdict=font1)
fig.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                   figsize=(7, 4))
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                   figsize=(7, 4))
someaxes["ax1"].scatter(x1, y1)  
someaxes["ax2"].bar(x2, y2)  
someaxes["ax3"].plot(x1, y1)   
plt.show() 
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                   figsize=(7, 4))
someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].bar(x2, y2)
someaxes["ax3"].plot(x1, y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                   figsize=(7, 4),
                                   layout='constrained')
                            
someaxes["ax1"].scatter(x1, y1)
someaxes["ax2"].bar(x2, y2)
someaxes["ax3"].plot(x1, y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
```

```python
#multiple viz on one axes object
#data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

#defining the figure and axes
fig, ax = plt.subplots(figsize=(7, 3))
#plotting two different visualizations on 
#the same axes object
ax.bar(x, y1, color="indigo")
ax.plot(x, y2, color="red")
```

```python
#add error information
y2_sd = np.std(y2)

#plotting
fig, ax = plt.subplots(figsize=(7, 3))
ax.bar(x, y1, color="red")

ax.errorbar(x, y2, yerr=y2_sd, fmt='none')

```

```python
#customizing error bars
fig, ax = plt.subplots(figsize=(7, 3))
ax.bar(x, y1, color="grey")
ax.errorbar(x, y2, yerr=y2_sd, fmt='none', 
            ecolor='indigo', elinewidth=3, 
            capsize=5, capthick=4)
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.bar(x, y1, color="grey")
ax.errorbar(x, y2, yerr=y2_sd, fmt="none", ecolor="indigo", 
            elinewidth=4, capsize=6, capthick=4, errorevery=2)

plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data=tips, # choose our dataset
x='total_bill', # define our x variable
y='tip') # define our y variable
```

```python
sns.set_style('whitegrid')
```

```python
tipgraph = sns.lineplot(data=tips,x='total_bill',y='tip')
tipgraph.set(title='Tips vs. Total Bill', xlabel='Total Bill ($)', ylabel='Tip Amount ($)')
```

```python
fig = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(data=tips,
x='total_bill',y='tip',
color = 'hotpink',
linestyle = '--',
linewidth = 3,
marker = 'o',
markerfacecolor = 'indigo')
```

```python
#adding multiple variables
tipgraph = sns.scatterplot(
    data=tips, 
    x='total_bill', 
    y='tip', 
    style='time', 
    hue='day', 
    palette=['purple', 'hotpink', 'deepskyblue', 'yellowgreen'])

tipgraph.set(
    title='Tips vs. Total Bill', 
    xlabel='Total Bill ($)', 
    ylabel='Tip Amount ($)')

plt.show()
```

```python
sns.pairplot(
data = tips,
hue = 'day')
```

```python
daysplot = sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    col="day",
    kind="scatter",
    col_wrap=2)

plt.show()
```

```python
#making a word cloud
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", on_bad_lines='skip')
df
```

```python
text = " ".join(each for each in df.quote) # join all our text from each row from our quote column into a string
# generate our wordcloud image
wordcloud = WordCloud(
    background_color="white", 
    colormap='inferno').generate(text)

fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")

plt.show()

```

```python
#Venn diagrams
#package for creating Venn diagrams in Python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry","apple", "key lime"])

diagram = venn2_unweighted(
    [A, B],
    set_labels=('Fruits', 'Pies'),
    set_colors=('blue', 'red'),
    alpha=0.5
)
plt.show()
```

```python
#customizing the labels

diagram = venn2_unweighted(
    [A, B],
    set_labels=('Fruits', 'Pies'),
    set_colors=('blue', 'red'),
    alpha=0.5)
diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))
plt.show()
```
